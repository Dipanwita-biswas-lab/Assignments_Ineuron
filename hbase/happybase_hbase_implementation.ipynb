```python
import happybase
```

```python
#hbase thrift start
```

```python
connection=happybase.Connection('ec2-54-221-171-130.compute-1.amazonaws.com')
```

```python
connection.open()
print('Connected to ', connection.host)
```

```python
# create a new table
#create ’tablename’, ‘columnfamily1’, ‘columnfamily2’
try:
    connection.create_table('airline',
                       {
                           'time': dict(),
                           'flight':dict()
                       })

    print('table created')
except Exception as e :
    print(e)
```

```python
#list tables
connection.tables()
```

```python
#is_enabled 'table 1'
connection.is_table_enabled('airline')
```

```python
#list columfamily
connection.table('airline').families()
```

```python
airline_table=connection.table('airline')
```

```python
# put data to the table
with open("AirLineData.csv") as data:
    lines= data.readlines()
    for line in lines: 
        line=line.strip()
        val= line.split(",")
        if val[22] =='0' and val[3]=='12':
            
            airline_table.put(val[0], 
                            {'time:year':val[1],
                            'time:month':val[2], 
                            'time:dayOfMonth':val[3], 
                            'flight:tailNum':val[11], 
                            'flight:origin':val[17],
                            'flight:dest':val[18]})
print('Data inserted')
```

```python
# scan the table 
#scan ‘customer’
for i in airline_table.scan():
    print(i)
```

```python
# delete rowkey rimi
airline_table.delete('ARY04F119')
```

```python
airline_table.regions()
```

```python
# display a particular row key
airline_table.row('ARY06F41506')
```

```python
#get ’table’ ‘rowkey’
for key, value in airline_table.row('ARY06F41506').items():
    print(key , ' = ', value)
```

```python
airline_table.rows('ARY06F41506')
```

```python
#disable ‘table’
connection.enable_table('airline')
```

```python
#disable ‘table’
connection.disable_table('airline')
# drop ’table’
connection.delete_table('airline')
print('table deletion sucessfull')
```

```python
#list tables
connection.tables()
```

```python
# close connection to the HBASe
connection.close()
print('connection closed')
```
